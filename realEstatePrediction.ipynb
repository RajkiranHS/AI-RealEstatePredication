Read and display the data

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

df = pd.read_csv('/content/homeprices.csv')
df
```

plot the data

```python
plt.xlabel('area')
plt.ylabel('price')
plt.scatter(df.area,df.price,color='red',marker='+')
plt.show()

```

Train the model and Predicate results




```python

model=LinearRegression()
model.fit(df[['area']],df.price)
model.predict([[3000]])

```

```python

pd.plotting.scatter_matrix(df)
plt.show()

plt.plot(df.area,model.predict(df[['area']]),color='blue')
plt.show()
df



pre=model.predict([[2600]])


newDf=df
newDf

inPreData=newDf[['area']]
inPreData

results=model.predict(inPreData)
results

newDf['results']=results
newDf

newDf.to_csv('HomePriceResults.csv')
newDf
newDf.rename(columns={'results':'pricePredication'})

plt.plot(df.area,model.predict(df[['area']]),color='blue')
plt.show()
df
```
